```python
import pandas as pd

from bs4 import BeautifulSoup
import requests
import urllib.request
from urllib.error import HTTPError
from urllib.error import URLError

from socket import timeout

import plotly.express as px
```

```python
df_sud = pd.read_csv('../raw_data/Sudan.csv',
                 header=1, index_col='Unnamed: 0')
df_sud
```

```python
df_alg = pd.read_csv('../raw_data/Algeria.csv',
                     header=1, index_col='Unnamed: 0')
df_alg
```

```python
df_alg['تاريخ يوم الواقعة'] = pd.to_datetime(df_alg['تاريخ يوم الواقعة'])
df_sud['يوم الواقعة'] = pd.to_datetime(df_sud['يوم الواقعة'])
start = '2018-12-1'
end = '2020-3-31'
mask = (df_alg['تاريخ يوم الواقعة'] >= start) & (df_alg['تاريخ يوم الواقعة'] <= end)
mask_s = (df_sud['يوم الواقعة'] >= start) & (df_sud['يوم الواقعة'] <= end)
alg_20 = df_alg.loc[mask]
sud_20 = df_sud.loc[mask_s]
uniq_alg = alg_20[alg_20['مصدر 1'].duplicated() == False]
uniq_sud = sud_20[sud_20['مصدر 1'].duplicated() == False]
```

```python
useless_alg = []
for x in uniq_alg['نص الخبر']:
    if len(str(x)) < 30 or x ==None:
        useless_alg.append(x)

useless_sud = []
for x in uniq_sud['نص الخبر']:
    if len(str(x)) < 30 or x ==None:
        useless_sud.append(x)
```

```python
print(f"Total links:      {alg_20['مصدر 1'].count()}")
print(f"unique:           {alg_20['مصدر 1'].nunique()}")
print(f"unique and empty: {len(useless_alg)}")
print('-------------')
print(f"Total links:      {sud_20['مصدر 1'].count()}")
print(f"unique:           {sud_20['مصدر 1'].nunique()}")
print(f"unique and empty: {len(useless_sud)}")
```

```python
df_alg['month'] = df_alg['تاريخ يوم الواقعة'].dt.month
df_sud['month'] = df_sud['يوم الواقعة'].dt.month
```

```python
test_df = df_alg[df_alg['نص الخبر'].duplicated() == False]
test_df = test_df.sample(10)
```

```python
links = [x for x in test_df['مصدر 1']]
links
```

```python
for i in range(len(links)):
    try:
        response = urllib.request.urlopen(links[i], timeout=10)
        print('Accessable')
    except HTTPError as error:
        print(error.reason)
    except URLError as error:
        if isinstance(error.reason, timeout):
            print(error.reason)
```

```python

def url_check(url):
    try:
        response = urllib.request.urlopen(url, timeout=100)
        return 'Accessable'
    except Exception as error:
        try:
            return error.reason
        except:
            return 'Not url'
```

```python
df_alg = df_alg.drop_duplicates()
df_alg.shape
```

```python
df_alg['link_status'] = df_alg['مصدر 1'].apply(url_check)
```

```python
df_alg.to_csv('../raw_data/Alg_updated.csv')
```

```python
df_sud['link_status'] = df_sud['مصدر 1'].apply(url_check)
```
